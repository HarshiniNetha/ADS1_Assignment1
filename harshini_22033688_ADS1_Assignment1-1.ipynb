```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load the data into a DataFrame

```python
df = pd.read_csv("https://www.metoffice.gov.uk/pub/data/weather/uk/climate/datasets/Rainfall/ranked/UK.txt", sep="\s+", skiprows=7, header=None)
```

Remove the last two columns (Index and Quality)

```python
df = df.iloc[:,:4]
```

Rename the columns

```python
df.columns = ["Year", "Rainfall (mm)", "Anomaly", "Station"]
```

Line plot

```python
df.plot(x="Year", y="Rainfall (mm)")
plt.title("Rainfall in the UK Over Time")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Year")
plt.show()
```

Bar chart

```python
df.plot.bar(x="Year", y="Rainfall (mm)")
plt.title("Rainfall in the UK by Year")
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.show()
```

Scatter plot

```python
df.plot.scatter(x="Year", y="Rainfall (mm)")
plt.title("Rainfall in the UK by Year")
plt.xlabel("Rainfall (mm)")
plt.ylabel("Year")
plt.show()
```
